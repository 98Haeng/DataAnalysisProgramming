```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('관광지별해시태그2.xlsx')
df.head()
```

```python
blog = pd.read_csv('관광지네이버블로그언급량.csv')
blog.head()
```

```python
df2 = pd.read_csv('충남관광지리스트.csv')
df2.head()
```

```python
len(df), len(blog), len(df2)
```

```python
final_df = pd.concat([df2, blog, df], axis = 1)
final_df.head()
```

```python
final_df['소거여부'].value_counts()
```

```python
final_df = final_df[final_df['소거여부']!='o']
final_df.head()
```

```python
final_df.drop('소거여부', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['가족여부'] = 0
final_df['연인여부'] = 0
final_df['친구여부'] = 0
final_df['아이여부'] = 0
final_df['휴식'] = 0
final_df['교육'] = 0
```

```python
final_df.dropna(axis=0, inplace=True)
final_df
```

```python
final_df.reset_index(drop=True, inplace=True)
```

```python
for i in range(len(final_df)):
    if '가족' in final_df.iloc[i]['해시태그']:
        final_df.at[i, '가족여부'] = 1
    if '연인' in final_df.iloc[i]['해시태그']:
        final_df.at[i, '연인여부'] = 1
    if '친구' in final_df.iloc[i]['해시태그']:
        final_df.at[i, '친구여부'] = 1
    if '아이' in final_df.iloc[i]['해시태그']:
        final_df.at[i, '아이여부'] = 1
    if '휴식' in final_df.iloc[i]['해시태그']:
        final_df.at[i, '휴식'] = 1
    if '교육' in final_df.iloc[i]['해시태그']:
        final_df.at[i, '교육'] = 1
```
