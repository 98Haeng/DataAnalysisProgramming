[크롤링 구조]
- url에 관광지를 입력해서 검색이 된다면, 바로 해시태그만 찾아와서 갖고오기
- 안된다면 우선 pass하고, 직접 들어가서 가져오는 크롤링 코드 작성

```python
from bs4 import BeautifulSoup
import time
import requests

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
from selenium.webdriver.common.action_chains import ActionChains
```

```python
# 데이터 불러오기
# 충청도 관광지 리스트
df = pd.read_csv('충남관광지리스트.csv')
df.head()
```

```python
df['해시태그'] = 0
```

```python
url_first = 'https://korean.visitkorea.or.kr/search/search_list.do?keyword='
search = '창덕궁'
url = url_first + search
driver = webdriver.Chrome(executable_path = 'C:/Users/chromedriver.exe') # 드라이버 열기
driver.implicitly_wait(3)
driver.get(url)
driver.fullscreen_window()
driver.set_window_size(1000, 3000)


tag_list = [] # 해시태그 리스트
for i in range(len(df)):
    search = df['관광지명'][i]
    url_first = 'https://korean.visitkorea.or.kr/search/search_list.do?keyword='
    url = url_first + search
    driver.get(url)
    
    time.sleep(2)
    
    driver.set_window_size(1000, 3000)
    # 1. 바로 나오는 경우 : 해시태그만 바로 갖고오기
    try:
        tags = driver.find_element(By.CLASS_NAME, 'tag')
        a = tags.text
        tags_col = a.replace('\n',', ')
        print(df['관광지명'][i] + '해시태그 완료')

    except:
        tags_col = '비어있음'
        print(df['관광지명'][i] + '해시태그 없음')
        
    finally:
        tag_list.append(tags_col)
        
    
```
