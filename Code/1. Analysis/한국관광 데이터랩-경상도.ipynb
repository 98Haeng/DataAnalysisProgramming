## 경상남도

```python
fn_2 = '경상남도 2022년 지역별 관광지 검색순위.xlsx'
df_2 = pd.read_excel(filepath + '/' + fn_2)
df_2.head()
```

```python
gsd = pd.concat([df_1,df_2])
gsd
```

```python
# 중분류 카테고리 비교 시각화
gw_middle = gsd.groupby(['중분류 카테고리'])[['검색건수']].count()
gw_middle.reset_index(drop=False, inplace=True)

gw_middle.sort_values(by='검색건수', ascending=False, inplace=True)
rcParams['figure.figsize'] = 12,7
sns.barplot(data = gw_middle,
           x=gw_middle['중분류 카테고리'],
           y=gw_middle['검색건수']);
plt.title('경상도 관광지 중분류 카테고리 비교', size=25)
plt.xticks(rotation=30,size=12)
plt.yticks(size=12)
plt.show()
```

```python
fn_3_1 = '경상북도 외지인 방문자 수_SKT.xlsx'
fn_3_2 = '경상북도 외지인 방문자 수_KT.xlsx'
visit1_1 = pd.read_excel(filepath + '/' + fn_3_1)
visit1_2 = pd.read_excel(filepath + '/' + fn_3_2)
```

```python
merge1 = pd.merge(visit1_1, visit1_2, on=['광역지자체명', '기초지자체명'])
merge1['기초지자체 방문자 수'] = merge1['기초지자체 방문자 수_x'] + merge1['기초지자체 방문자 수_y']
merge1 = merge1[['광역지자체명', '기초지자체명', '기초지자체 방문자 수']]

# 경상북도의 기초지자체 방문자 수 합 구하기
total1 = merge1['기초지자체 방문자 수'].sum()

# 새로운 행 생성
new_row = pd.DataFrame({'광역지자체명': '합계', '기초지자체명': ':', '기초지자체 방문자 수': total1}, index=[len(merge1)])

# 총 방문자수를 포함한 데이터프레임
visitors1 = pd.concat([merge1, new_row], axis=0, ignore_index=True)
visitors1
```

```python
merge1.sort_values(by='기초지자체 방문자 수', ascending=False, inplace=True)
rcParams['figure.figsize'] = 12,7
sns.barplot(data = merge1,
           x=merge1['기초지자체명'],
           y=merge1['기초지자체 방문자 수']);
plt.title('경북 기초지자체별 외지인 방문자 수 비교', size=25)
plt.xticks(rotation=30,size=12)
plt.yticks(size=12)
plt.show()
```

```python
fn_4_1 = '경상남도 외지인 방문자 수_SKT.xlsx'
fn_4_2 = '경상남도 외지인 방문자 수_KT.xlsx'
visit2_1 = pd.read_excel(filepath + '/' + fn_4_1)
visit2_2 = pd.read_excel(filepath + '/' + fn_4_2)
```

```python
merge2 = pd.merge(visit2_1, visit2_2, on=['광역지자체명', '기초지자체명'])
merge2['기초지자체 방문자 수'] = merge2['기초지자체 방문자 수_x'] + merge2['기초지자체 방문자 수_y']
merge2 = merge2[['광역지자체명', '기초지자체명', '기초지자체 방문자 수']]

# 경상남도의 기초지자체 방문자 수 합 구하기
total2= merge2['기초지자체 방문자 수'].sum()

# 새로운 행 생성
new_row = pd.DataFrame({'광역지자체명': '합계', '기초지자체명': ':', '기초지자체 방문자 수': total2}, index=[len(merge2)])

# 총 방문자수를 포함한 데이터프레임
visitors2 = pd.concat([merge2, new_row], axis=0, ignore_index=True)
visitors2
```

```python
merge2.sort_values(by='기초지자체 방문자 수', ascending=False, inplace=True)
rcParams['figure.figsize'] = 12,7
sns.barplot(data = merge2,
           x=merge2['기초지자체명'],
           y=merge2['기초지자체 방문자 수']);
plt.title('경남 기초지자체별 외지인 방문자 수 비교', size=25)
plt.xticks(rotation=30,size=12)
plt.yticks(size=12)
plt.show()
```

```python
gsd_total_visitors = pd.concat([merge1, merge2], axis=0, ignore_index=True)

# 기초지자체 방문자 수 합 구하기
total= gsd_total_visitors['기초지자체 방문자 수'].sum()

# 새로운 행 생성
new_sum = pd.DataFrame({'광역지자체명': '합계', '기초지자체명': ':', '기초지자체 방문자 수': total}, index=[len(gsd_total_visitors)])

# 경상도 총 방문자수를 포함한 데이터프레임
gsd_total_visitors = pd.concat([gsd_total_visitors, new_sum], axis=0, ignore_index=True)

gsd_total_visitors
```
