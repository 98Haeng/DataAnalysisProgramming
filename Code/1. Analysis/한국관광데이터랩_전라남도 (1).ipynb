# 기초지자체

```python
#기초지자체 검색 건수
sch = pd.read_csv('/Users/jeongseung-a/Desktop/데분프 플젝/전라남도_지역별 검색건수.csv',encoding='cp949')
```

```python
c = sch.sort_values(by="기초지자체 검색건수",ascending = False ,inplace = False)
c.head()
```

```python
#기초지자체(시/군/구) 검색건수
spot = pd.read_excel('/Users/jeongseung-a/Desktop/데분프 플젝/전라남도_관광지 검색순위.xlsx')
jns = spot.groupby(['시/군/구'])[['검색건수']].count()
jns.reset_index(drop=False, inplace=True)
jns.sort_values(by="검색건수",ascending = False ,inplace = False)
```

```python
jns.sort_values(by='검색건수', ascending=False, inplace=True)
rcParams['figure.figsize'] = 12,7
sns.barplot(data = jns,
           x=jns['시/군/구'],
           y=jns['검색건수']);
plt.title('전라남도 지역별 검색건수 비교', size=25)
plt.xticks(rotation=30,size=12)
plt.yticks(size=12)
plt.show()
```
