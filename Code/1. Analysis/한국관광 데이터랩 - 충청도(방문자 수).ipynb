```python
# 라이브러리
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from matplotlib import rcParams
%matplotlib inline
import warnings
warnings.filterwarnings('ignore') # 경고 무시
import math

from matplotlib import rc
rc('font', family='gulim')
plt.rcParams['axes.unicode_minus'] = False

sns.set_palette('pastel')
```

# 충청북도 방문자 수

```python
filepath1 = '/Users/qsc02/desktop/학교 자료/데분프/팀플/data/충청북도 2022년 방문자 수'
filelist1 = os.listdir(filepath1)
filelist1
```

```python
cb_1 = pd.read_excel(filepath1 + '/' + '충청북도 방문자 수 추이.xlsx')
cb_2 = pd.read_excel(filepath1 + '/' + '충청북도 지역별 방문자 수.xlsx')
```

```python
cb1 = cb_1.set_index('방문자 구분').loc['외부방문자(b+c)']
cb1['방문자 수'] = round(cb1['방문자 수']/10000,4)
rcParams['figure.figsize'] = 12,7
sns.barplot(data = cb1,
           x=cb1['기준년월'],
           y=cb1['방문자 수']);
plt.title('충청북도 월별 외부방문자 수', size=20)
plt.xticks(rotation=30,size=12)
plt.yticks(size=12)
plt.show()
```

```python
cb2 = cb_2.groupby('기초지자체명')[['기초지자체 방문자 수']].sum()
cb2['기초지자체 방문자 수'] = round(cb2['기초지자체 방문자 수']/10000,4)
rcParams['figure.figsize'] = 12,7
sns.barplot(data = cb2,
           x=cb2.index,
           y=cb2['기초지자체 방문자 수']);
plt.title('충청북도 지자체별 방문자 수', size=20)
plt.xticks(rotation=30,size=12)
plt.yticks(size=12)
plt.show()
```
