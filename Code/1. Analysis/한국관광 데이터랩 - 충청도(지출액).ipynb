```python
# 라이브러리
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from matplotlib import rcParams
%matplotlib inline
import warnings
warnings.filterwarnings('ignore') # 경고 무시
import math

from matplotlib import rc
rc('font', family='gulim')
plt.rcParams['axes.unicode_minus'] = False

sns.set_palette('pastel')
```

# 충청북도 지출액

```python
filepath1 = '/Users/qsc02/desktop/학교 자료/데분프/팀플/data/충청북도 2022년 지출액'
filelist1 = os.listdir(filepath1)
filelist1
```

```python
cb_1 = pd.read_excel(filepath1 + '/' + '충청북도 내국인 관광소비 추이.xlsx')
cb_2 = pd.read_excel(filepath1 + '/' + '충청북도 내국인 업종별 지출액.xlsx')
cb_3 = pd.read_excel(filepath1 + '/' + '충청북도 내국인 지역별 지출액.xlsx')
```

```python
cb_f_1 = pd.read_excel(filepath1 + '/' + '충청북도 외국인 관광소비 추이.xlsx')
cb_f_2 = pd.read_excel(filepath1 + '/' + '충청북도 외국인 업종별 지출액.xlsx')
cb_f_3 = pd.read_excel(filepath1 + '/' + '충청북도 외국인 지역별 지출액.xlsx')
```

```python
cb1 = cb_1.set_index('중분류').loc['관광총소비']
cb1['지출액'] = round(cb1['지출액']/1000,2)
rcParams['figure.figsize'] = 12,7
sns.barplot(data = cb1,
           x=cb1['기준년월'],
           y=cb1['지출액']);
plt.title('충청북도 월별 지출액', size=20)
plt.xticks(rotation=30,size=12)
plt.yticks(size=12)
plt.show()
```

```python
cb2 = cb_2.groupby('중분류')[['중분류 지출액']].sum()
cb2['중분류 지출액'] = round(cb2['중분류 지출액']/1000,2)
rcParams['figure.figsize'] = 12,7
sns.barplot(data = cb2,
           x=cb2.index,
           y=cb2['중분류 지출액']);
plt.title('충청북도 중분류별 지출액', size=20)
plt.xticks(rotation=30,size=12)
plt.yticks(size=12)
plt.show()
```

```python
cb3 = cb_3.groupby('기초지자체 명')[['기초지자체 지출액']].sum()
cb3['기초지자체 지출액'] = round(cb3['기초지자체 지출액']/1000,2)
rcParams['figure.figsize'] = 12,7
sns.barplot(data = cb3,
           x=cb3.index,
           y=cb3['기초지자체 지출액']);
plt.title('충청북도 지자체별 지출액', size=20)
plt.xticks(rotation=30,size=12)
plt.yticks(size=12)
plt.show()
```
