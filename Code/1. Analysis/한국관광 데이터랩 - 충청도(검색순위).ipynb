```python
# 라이브러리
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from matplotlib import rcParams
%matplotlib inline
import warnings
warnings.filterwarnings('ignore') # 경고 무시

from matplotlib import rc
rc('font', family='gulim')
plt.rcParams['axes.unicode_minus'] = False

sns.set_palette('pastel')
```

```python
# 네비게이션 데이터 경로 설정
filepath = '/Users/qsc02/desktop/학교 자료/데분프/팀플/data'
filelist = os.listdir(filepath)
filelist
```

# 충청북도

```python
fn_1 = '충청북도 2022년 지역별 관광지 검색순위.xlsx'
df_1 = pd.read_excel(filepath + '/' + fn_1)
df_1.head()
```

```python
# 충청북도 중분류 카테고리 비교 시각화
gw_middle_1 = df_1.groupby(['중분류 카테고리'])[['검색건수']].count()
gw_middle_1.reset_index(drop=False, inplace=True)
```

```python
gw_middle_1.sort_values(by='검색건수', ascending=False, inplace=True)
rcParams['figure.figsize'] = 12,7
sns.barplot(data = gw_middle_1,
           x=gw_middle_1['중분류 카테고리'],
           y=gw_middle_1['검색건수']);
plt.title('충청북도 관광지 중분류 카테고리 비교', size=25)
plt.xticks(rotation=30,size=12)
plt.yticks(size=12)
plt.show()
```

```python
sgg_1 = df_1.groupby(['시/군/구'])[['검색건수']].count()
sgg_1.reset_index(drop=False, inplace=True)
sgg_1
```

```python
sgg_1.sort_values(by='검색건수', ascending=False, inplace=True)
rcParams['figure.figsize'] = 12,7
sns.barplot(data = sgg_1,
           x=sgg_1['시/군/구'],
           y=sgg_1['검색건수']);
plt.title('충청북도 지역별 관광지 수 비교', size=25)
plt.xticks(rotation=30,size=12)
plt.yticks(size=12)
plt.show()
```
